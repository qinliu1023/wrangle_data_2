### Fixing Unexpected Street Types

```python
"""
Fixing unexpected street types to the appropriate ones in the expected list.

The update_name function fixes the street name. The function takes a string 
with street name as an argument and would return the fixed name
"""

import xml.etree.cElementTree as ET
from collections import defaultdict
import re
import pprint

street_type_re = re.compile(r'\b\S+\.?$', re.IGNORECASE)


expected = ["Street", "Avenue", "Boulevard", "Drive", "Court", "Place", "Square", "Lane", "Road", 
            "Trail", "Parkway", "Commons"]

mapping = { "St": "Street",
            "St.": "Street",
            "Ave": "Avenue",
            "Rd.": "Road"
            }


def audit_street_type(street_types, street_name):
    m = street_type_re.search(street_name)
    if m:
        street_type = m.group()
        if street_type not in expected:
            street_types[street_type].add(street_name)


def is_street_name(elem):
    return (elem.attrib['k'] == "addr:street")


def audit(OSM_PATH):
    osm_file = open(OSM_PATH, "r")
    street_types = defaultdict(set)
    for event, elem in ET.iterparse(osm_file, events=("start",)):
        if elem.tag == "node" or elem.tag == "way":
            for tag in elem.iter("tag"):
                if is_street_name(tag):
                    audit_street_type(street_types, tag.attrib['v'])
    osm_file.close()
    return street_types


def update_name(name, mapping):
    m = street_type_re.search(name)
    if m:
        street_type = m.group()
        if street_type not in expected:
            name = re.sub(street_type_re, mapping[street_type], name)

    return name
```

## Additional Ideas 

### Data Improment Ideas

    When I submit queries looking for amenities there is no convenience markets like shoprite and walmart comes out. However, when I using filter UPPER(value) = 'SHOPRITE' to do the query again, there do come out some records. 
    
    When I look into this data, the most common key for value shoprite is name which is not that reasonable. And when I look for walmart, there is no outcomes.
    
    Anticipated Issues:
    1. When someone wants to do a research about coverage of markets/supermarkes of new york area and make a decision about where to open a new branch of the market/supermarket, the data might be too dirty and not sufficient for analysis and the cleaning process would be much more difficult. And this means questions this data can answer might be limited. 
    
    Here comes my solution:
    1. Standardize keys for certain category values, for example, a supermarkets could be assigned a key as amenity
    2. When a user contributes to the data a similar has already existed in the data, the existing key and suggested key should be prompted out to help the new contributor to better improve the data
    2. Encourage users not only contribute to common ones but also to some which can easily be forgotten by people, like supermarkets here.
    
    If above issue could be solved, then the data would be much more sufficient to answer more various answers with less efforts, which is an increase in work efficiency.
